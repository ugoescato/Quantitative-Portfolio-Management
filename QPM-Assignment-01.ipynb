```python
import yfinance as yf
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

### Question 1.1

```python
tickers = ["META", "AMZN", "AAPL", "NFLX", "GOOG"]
start_date = "2015-01-01"
end_date = "2020-12-31"

faang_data = yf.download(tickers, start=start_date, end=end_date)
faang_close = faang_data['Adj Close']
prices=faang_close
prices
```

### Question 1.2

```python
# Mean of daily returns
returns = prices.pct_change().dropna()
returns.mean()
```

```python
# Variance of daily returns
returns.std()**(2)
```

```python
# Covariance of daily returns
returns.cov()
```

### Question 1.3

```python
# Skewness of daily returns distributions
returns.skew()
```

```python
# Excess kurtosis of daily returns distributions
returns.kurt()
```

Skewness of daily returns is non-null and excess kurtosis is positive, hence daily returns are not normally distributed.

### Question 2.1

```python
# Downloading S&P 500 table 
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
tables = pd.read_html(url)
SP500_table = tables[0]
SP500_table
```

```python
# Tickers
tickers_SP500 = SP500_table["Symbol"]
tickers_SP500
```

```python
start_date = "2000-01-01"
end_date = "2022-12-31"
```

```python
tickers_SP500 = tickers_SP500.tolist()

SP_data = yf.download(tickers_SP500, start=start_date, end=end_date)
SP_close = SP_data['Adj Close']
prices_SP500 = SP_close

prices_SP500.head()
```

```python
# Drop NA's
prices_SP500 = prices_SP500.dropna(axis=1, how='all')
prices_SP500.head()
```

```python
# Drop also company names with more than 100 missing observations
number_of_null = prices_SP500.isnull().sum() # This shows how many null values per ticker
print(number_of_null)
```

```python
prices_SP500 = prices_SP500.loc[:,number_of_null<100]
((prices_SP500.isnull().sum())>=100).sum() #counts number of columns with more than 100 null values, should display 0
```

```python
prices_SP500.head()
```

### Question 2.2

```python
log_returns = np.log(prices_SP500 / prices_SP500.shift(1))
log_returns
```

### Question 2.3

```python
# Annualized Mean Return
annualized_mean_return = log_returns.mean()*252
print(annualized_mean_return)
```

```python
# Annualized Volatility 
annualized_volatility = log_returns.std()*np.sqrt(252)
print(annualized_volatility)
```

```python
data = yf.download('^TNX', start=start_date, end=end_date) # U.S. 10-Year Treasury Bond yield as a proxy for rf on this period
data = data['Adj Close'] / 100
rf = np.array(data.mean())
print(rf)
```

```python
sharpe_ratios =  (annualized_mean_return - rf) / annualized_volatility
print(sharpe_ratios)
```

### Question 2.4

```python
weights = sharpe_ratios / sharpe_ratios.sum()
print(weights)
```

Relying solely on the Sharpe ratio of individual stocks in our dataset to determine portfolio weights is inadequate because it overlooks the covariance between stocks. This means we might miss valuable diversification benefits that come from combining assets with different movement patterns, even if some have lower Sharpe ratios. Additionally, historical Sharpe ratios can carry significant estimation errors, particularly in the mean returns component, making them unreliable indicators of future risk-adjusted performance if used alone. Furthermore, investors may have varying risk tolerances and return expectations that the Sharpe ratio does not capture (example: Maximum Drawdown, Time under Water and so on). Lastly, past performance is not a guarantee of future results.
